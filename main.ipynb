```python
df2=df[List_columns].copy()
```

In this example we want to distribute the prices into 3 categories A,B,C and D and with A the highest range of prices.

In order to chose the price ranges we will be based on the grouping by grades and find the max price within each grade

```python
df2[['grade','price']].groupby(['grade']).agg(['max']).round(1).reset_index()
```

We will chose our 4 categories based on price range from grade 1 to grade 4(We saw in the first example that grade 5 is a sort of a boundaries between 2 luxury level).
Then, based on the average prices for every grade, we can notice that between grade 9 and grade 10 there is a big gap between the everage and there is not intersection looking at the maximums. So we will choose the ranges as following:

```python
df['price'].min() # minimum price to start from
```

Category D: Between 75000 and 435000 (max price from grade 4)

Category C: Between 435000 and 795000 (No houses lower than grade 5)

Category B: Between 795000 and 1200000 (No houses lower than grade 6)

Category A: Higher than 1200000 (No houses lower than grade 7)

```python
def condition(x):
    if (x>=75000 and x<435000): 
        return 'D' 
    elif (x>=435000 and x<795000): 
        return 'C'
    elif (x>=795000 and x<1200000):
        return 'B'
    else:
         return 'A'
```

```python
df2['Price_Category']=df2['price'].apply(condition)
```
